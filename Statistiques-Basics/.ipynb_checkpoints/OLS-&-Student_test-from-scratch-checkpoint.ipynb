```python
import numpy as np
import pandas as pd
from numpy.linalg import *
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.pyplot import figure
from mpl_toolkits.mplot3d import Axes3D
import sklearn.linear_model as lm
from sklearn.linear_model import LinearRegression as lr
```

## 1- Afficher les 5 premières lignes du Dataframe :

```python
data = pd.read_csv(url, sep=" ", header='infer')
data.columns = ["year","gnp","invest","cpi","interest"]
print (data.head(5))
```

## 2- Réaliser le graph :

```python
X = np.log(data['gnp'])
Y = np.log(data['invest'])

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(X,Y,'o', label="Data")
plt.ylabel('Investment', fontsize=18)
plt.xlabel('GNP', fontsize=18)
plt.title("Investment function of GNP")
plt.legend()
plt.show()
```

## 3- Pente, Intercept, ecart-type et coefficiant de détermination

#### Pour toute la suite, B0 et B1 correspondent respectivement aux estimateurs de $θ_{0}$ et $θ_1$ dans l'équation type : $Y$ = $θ_{0}$ + $X^T$*$θ_1$ + $ε$

### Calcul de Xmean et Ymean

```python
n = np.size(X) # est aussi égale à np.size(Y) est égale au nombre de lignes (ou observations)
Xmean = np.sum(X)/n
Ymean = np.sum(Y)/n
print ("Xmean = " + str(Xmean), "\nYmean = " + str(Ymean))
```

### Calcul de pente B1 et de l'intercept B0

```python
B1 = np.sum((X-Xmean)*(Y-Ymean))/np.sum((X-Xmean)**2) 
print ("Pente: B1 = " + str(B1))

B0 = Ymean - B1*Xmean 
print ("Intercept: B0 = "+ str(B0))
```

### Calcul de σ*², l'estimateur sans biais de la variance du bruit ε

```python
Ychap = B0+B1*X
sigma2 = (1/(n-2))*np.sum((Y-Ychap)**2) 
print ("σ*² = " + str(sigma2))
```

### Calcul des écart-types de B0 et B1 et coefficient de détermination R²

```python
# Calcul de V[B0] la variance de B0 et S[B0] son écart-type
VB0 = sigma2*((1/n)+(Xmean**2/(np.sum((X-Xmean)**2))))
SB0 = np.sqrt(VB0)
print ("écart-type: S[B0] = " + str(SB0))

# Calcul de V[B1] la variance de B1 et S[B1] son écart-type
VB1 = sigma2/np.sum((X-Xmean)**2) 
SB1 = np.sqrt(VB1)
print ("\nécart-type: S[B1] = " + str(SB1))

# Calcul de R² le coefficient de détermination
R2 = np.sum((Ychap-Ymean)**2)/np.sum((Y-Ymean)**2)
print ("\nR² = " + str(R2))
```

## 4- Test de Student sur la pente B1 

##### On veut réaliser un test de Student et vérifier l'hypothèse nulle $H_0$ : la pente B1 = 0

```python
# Soit T la statistique de test: 
# Dans le cas de l'hypothèse nulle on prend teta* égal à zéro
T = (B1-0)/SB1
print ("t-stat T = " + str(T))

# La p-value est le risque de première espèce le plus petit correspondant au t-stat:
pval = (1 - stats.t.cdf(abs(T),df = n-2))*2
print ("\np-value = " + str(pval))
```

##### p-value < 0.01 par conséquent on rejette l'hypothèse $H_0$

### Pour la suite nous allons passer en matriciel pour simplifier les calculs. Nous allons définir de nouvelles matrices Xnew : [ 1n, X ] et Bn : [ B0, B1 ]

```python
Xcol = np.array(X)
ones = np.ones(len(X))
Xnew_T = np.vstack((ones,Xcol))
Xnew = Xnew_T.transpose()
print ("Xnew = \n" + str(Xnew))
print ("Dim(Xnew) : " + str(Xnew.shape))

Bn_T = np.array([B0,B1])
Bn = Bn_T.transpose()
print ("\nBn = " + str(Bn))
print ("Dim(Bn) : " + str(Bn.shape))
```

## 5- Prédiction et IC

```python
# Pour GNP = 1000, l'investissement prédis (Y_1000) vaut:
Ychap_1000 = np.exp(B0 + np.log(1000)*B1)
print ("Investissement prédis: Ychap_1000 = " + str(Ychap_1000))

# α = 10% --> α/2 = 5% --> 1-α/2 = 95%
Q = stats.t.ppf(0.95,n-2)

x_T = np.array([1,np.log(1000)])
x = x_T.transpose()

CI_sup = np.exp(x_T.dot(Bn) + Q*np.sqrt(sigma2)*np.sqrt(x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x)))
CI_inf = np.exp(x_T.dot(Bn) - Q*np.sqrt(sigma2)*np.sqrt(x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x)))

PI_sup = np.exp(x_T.dot(Bn) + Q*np.sqrt(sigma2)*np.sqrt(1 + x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x)))
PI_inf = np.exp(x_T.dot(Bn) - Q*np.sqrt(sigma2)*np.sqrt(1 + x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x)))
print ("\n(echelle naturelle) CI(x=1000) : [" + str(CI_inf) + " < Ychap_1000 < " + str(CI_sup) + "]") 
print ("(echelle naturelle) PI(x=1000) : [" + str(PI_inf) + " < Y_1000 < " + str(PI_sup) + "]") 


CI_sup_log = x_T.dot(Bn) + Q*np.sqrt(sigma2)*np.sqrt(x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x))
CI_inf_log = x_T.dot(Bn) - Q*np.sqrt(sigma2)*np.sqrt(x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x))

PI_sup_log = x_T.dot(Bn) + Q*np.sqrt(sigma2)*np.sqrt(1 + x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x))
PI_inf_log = x_T.dot(Bn) - Q*np.sqrt(sigma2)*np.sqrt(1 + x_T.dot(inv(Xnew_T.dot(Xnew))).dot(x))
print ("\n(echelle log) CI(x=log(1000)) : [" + str(CI_inf_log) + " < Ychap_1000 < " + str(CI_sup_log) + "]") 
print ("(echelle log) PI(x=log(1000)) : [" + str(PI_inf_log) + " < Y_1000 < " + str(PI_sup_log) + "]") 
```

## 6- Graph avec Régression, CI et PI 

```python
graph_CI_sup = []
graph_CI_inf = []
graph_PI_sup = []
graph_PI_inf = []

for i in X:
    xi_T = np.array([1,i])
    xi = xi_T.transpose()
    CI_sup = xi_T.dot(Bn) + Q*np.sqrt(sigma2)*np.sqrt(xi_T.dot(inv(Xnew_T.dot(Xnew))).dot(xi))
    CI_inf = xi_T.dot(Bn) - Q*np.sqrt(sigma2)*np.sqrt(xi_T.dot(inv(Xnew_T.dot(Xnew))).dot(xi))
    PI_sup = xi_T.dot(Bn) + Q*np.sqrt(sigma2)*np.sqrt(1 + xi_T.dot(inv(Xnew_T.dot(Xnew))).dot(xi))
    PI_inf = xi_T.dot(Bn) - Q*np.sqrt(sigma2)*np.sqrt(1 + xi_T.dot(inv(Xnew_T.dot(Xnew))).dot(xi))
    graph_CI_sup.append(CI_sup)
    graph_CI_inf.append(CI_inf)
    graph_PI_sup.append(PI_sup)
    graph_PI_inf.append(PI_inf)

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(X,Y,'o', label="Data",linewidth=1)
plt.plot(X,Ychap, label="Regression",linewidth=1)
plt.plot(X,graph_CI_sup, label="CI_sup",linewidth=1,color = "red")
plt.plot(X,graph_CI_inf, label="CI_inf",linewidth=1,color = "red")
plt.plot(X,graph_PI_sup, label="PI_sup",linewidth=1,color = "green")
plt.plot(X,graph_PI_inf, label="PI_inf",linewidth=1,color = "green")
plt.ylabel('Investment', fontsize=18)
plt.xlabel('GNP', fontsize=18)
plt.title("Investment function of GNP")
plt.legend()
plt.show()
```

## 9- Ajout de la variable Interest

```python
Y2 = Y
X2_T = np.vstack((X,data['interest']))
X2 = X2_T.transpose()
X2 = sm.add_constant(X2)
X2_T = X2.transpose()

print ("Matrice X de l'équation normale : \n" + str(X2))

gram_matrix = X2_T.dot(X2)
print ("\nMatrice de Gram : \n" + str(gram_matrix))
print ("\nRang de la matrice de Gram : " + str(np.linalg.matrix_rank(gram_matrix)))
```

##### Le rang de la matrice de Gram vaut 3, elle donc de rang plein et donc INVERSIBLE

## 10- Calcul des coefficients, écart-types, coef. de détermination et test de Student

### Calcul des coefficients de régression $B_n$ $[B_0$, $B_1$, $B_2]$

##### Dans cette partie nous allons calculer les coefficients de régression à partir de la matrice de Gram car elle est inversible. Nous allons aussi calculer ses valeurs propres afin de discuter de la significativité des coefficients plus tard.

```python
Bn2 = inv(X2_T.dot(X2)).dot(X2_T).dot(Y)
B_0 = Bn2[[0]]
B_1 = Bn2[[1]]
B_2 = Bn2[[2]]
print ("Les 3 coefficients de la regressions sont: \nB_0 = "+ str(B_0))
print ("B_1 = "+ str(B_1))
print ("B_2 = "+ str(B_2))

print ("\nvaleurs propres de la matrice de Gram : \n" + str(np.linalg.eigvals(gram_matrix)))
```

### Calcul de R² et σ*²

```python
Y2chap = X2 @ Bn2 
Y2mean = Ymean

# Calcul de R² le coefficient de détermination
R_2 = np.sum((Y2chap-Y2mean)**2)/np.sum((Y2-Y2mean)**2)
print ("\n1ère méthode: R² = " + str(R_2))

# Calcul de σ*², l'estimateur sans biais de la variance du bruit ε
sigma_2 = (1/(n-3))*np.sum((Y2-Y2chap)**2) 
print ("\nσ*² = " + str(sigma_2))
```

##### Afin de déterminer les écarts-types des coefficients de régression, nous allons calculer la matrice de covariance de B2n

```python
# Calcul du la matrice de covariance Cov(B2n): 
covB2n = sigma_2*inv(X2_T @ X2)

# La diagonale de la matrice contient les variances:
VB_0 = np.diag(covB2n)[0]
VB_1 = np.diag(covB2n)[1]
VB_2 = np.diag(covB2n)[2]

# Calcul des écarts-types à partir des variances:
SB_0 = np.sqrt(np.diag(covB2n)[0])
SB_1 = np.sqrt(np.diag(covB2n)[1])
SB_2 = np.sqrt(np.diag(covB2n)[2])

print ("Matrice de covariance : \n" + str(covB2n))
print ("\nEcart-type de B_0 : SB_0 = " + str(SB_0))
print ("\nEcart-type de B_1 : SB_1 = " + str(SB_1))
print ("\nEcart-type de B_2 : SB_2 = " + str(SB_2))
```

### Test de Student avec Hypothèse $H_0$ : $B_i$ = 0

```python
# Soit T la statistique de test: 
T0 = (B_0-0)/SB_0
print ("\nt-stat T(B_0) = " + str(T0))

# La p-value est le risque de première espèce le plus petit correspondant au t-stat:
pval0 = (1 - stats.t.cdf(abs(T0),df = n-3))*2
print ("p-value de B_0 = " + str(pval0))

T1 = (B_1-0)/SB_1
print ("\nt-stat T(B_1) = " + str(T1))

pval1 = (1 - stats.t.cdf(abs(T1),df = n-3))*2
print ("p-value de B_1 = " + str(pval1))

T2 = (B_2-0)/SB_2
print ("\nt-stat T(B_2) = " + str(T2))

pval2 = (1 - stats.t.cdf(abs(T2),df = n-3))*2
print ("p-value de B_2 = " + str(pval2))
```

##### On constate que la p-value de B_2 >> 0,1 . Par conséquent il n'y a pas de présomption contre l'hypothèse nulle dans ce cas. Pour les autres coefficients, l'hypothèse nulle est rejetée. On en conclura que le coefficient $B_2$ n'est pas significatif. On aurait pu s'en douter, car comme vu précédemment, la matrice de Gram possède une valeur propore très proche de zéro.
